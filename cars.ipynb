```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(50, 20))

n = len(df)
mcons, msp, mcyl, mhp = [],[],[],[]

m1 = numpy.mean(df['Consumption'])
m2 = numpy.mean(df['Speed'])
m3 = numpy.mean(df['Cyl'])
m4 = numpy.mean(df['Hp'])

for i in range(0, n):
    mcons.append(m1)
    msp.append(m2)
    mcyl.append(m3)
    mhp.append(m4)

plt.subplot(2, 2, 1)
plt.bar(df['Car'], df['Consumption'])
# plt.scatter(x=range(0,n), y=mcons)
plt.plot(mcons, 'g', linewidth=5)


plt.subplot(2, 2, 2)
plt.bar(df['Car'], df['Speed'])
plt.plot(msp, 'pink', linewidth=12)

plt.subplot(2, 2, 3)
plt.bar(df['Car'], df['Cyl'])
plt.plot(mcyl, 'magenta', linewidth=8)

plt.subplot(2, 2, 4)
plt.bar(df['Car'], df['Hp'])
plt.plot(mhp, '-r', linewidth=2)

plt.show()
```
